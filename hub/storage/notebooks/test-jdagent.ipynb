# Test JD Agent Python Package

Testing the `jdagent` package for hub, claude, and vault access.

```python
# Add jdagent package to path
import sys
sys.path.insert(0, '/Users/jddavenport/Projects/JD Agent/hub/notebooks')

# Import jdagent
from jdagent import hub, claude, vault
print('jdagent package imported successfully!')
```

## Test Hub Connection

```python
# Check hub connection
print(f'Hub connected: {hub.is_connected()}')
print(f'Health: {hub.health()}')
```

## Test Tasks API

```python
# Get tasks
tasks = hub.tasks.list(limit=5)
print(f'Found {len(tasks)} tasks')
for t in tasks[:3]:
    print(f"  - {t.get('title', 'Untitled')} [{t.get('status')}]")
```

## Test Projects API

```python
# Get projects
projects = hub.projects.list()
print(f'Found {len(projects)} projects')
for p in projects[:3]:
    print(f"  - {p.get('name', 'Untitled')} [{p.get('status')}]")
```

## Test Vault Search

```python
# Search vault
results = vault.search('notebook', limit=5)
print(f'Found {len(results)} vault entries matching "notebook"')
for r in results[:3]:
    print(f"  - {r.get('title', 'Untitled')} [{r.get('contentType')}]")
```

## Test Vault Stats

```python
# Get vault stats
stats = vault.stats()
print('Vault Statistics:')
for key, value in stats.items():
    print(f'  {key}: {value}')
```
